```python
#图像地址和该py文件在同一文件夹下,否则需要修改路径
filename = 'tomandjerry.png'
#读取图像
img = cv.imread(filename, 0)

 
#傅里叶变换
f = np.fft.fft2(img)
# 频谱中心化
fshift = np.fft.fftshift(f)
# 计算频谱
res = np.log(np.abs(fshift))
# tmpRes 用于显示过滤区域
tmpRes = copy.copy(res)

# 高通滤波
# 过滤区域是频谱的中心区域filterWidth为边长的正方形
# 过滤区域越大, 过滤掉的低频成分越多, 背景黑化越严重
filterWidth = 25
rows, cols = img.shape
crow, ccol = int(rows/2), int(cols/2)
# 频谱中,中心白亮的区域是低频区, 让频谱的低频区等与0,即低频被过滤掉
# 只有高频通过, 所以叫高频滤波
# 低频是背景部分, 背景被过滤, 用黑色填充, 逆变换后生成的图片
# 应该是黑色背景, 白色轮廓的黑白图像
fshift[crow-filterWidth:crow+filterWidth, ccol-filterWidth:ccol+filterWidth]=0

# 用于显示过滤区域
tmpRes[crow-filterWidth:crow+filterWidth, ccol-filterWidth:ccol+filterWidth]=0


# 频谱中心逆变换
ishift = np.fft.ifftshift(fshift)
#傅里叶逆变
iimg = np.fft.ifft2(ishift)
iimg = np.abs(iimg)


# 绘制滤波前的频谱
plt.subplot(141), plt.imshow(res, 'gray'), plt.title('frequency content before filtering')
plt.axis('off')
# # 绘制滤波后的频谱
plt.subplot(142), plt.imshow(tmpRes, 'gray'), plt.title('filter area') 
plt.axis('off')
# 绘制原始图像
plt.subplot(143), plt.imshow(img, 'gray'), plt.title('Original Image')
# 绘制滤波后的图像
plt.subplot(144), plt.imshow(iimg, 'gray'), plt.title('Height Pass Image')
plt.axis('off')
plt.show()
```

```python
cap = cv.VideoCapture('tomandjerry.blv')
# 
fourcc = cv.VideoWriter_fourcc(*'XVID')

filterWidth = 25
height = 360; width = 480
# 不加False会错
# 24帧通过计算得到, 加下面注解
out = cv.VideoWriter('tomandjerry.mp4', fourcc, 24.0, (cols,rows), False)
while(cap.isOpened()):
    ret, frame = cap.read()##ret返回布尔量
    if not ret:
        break
    
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    #傅里叶变换
    img = gray
    
    f= np.fft.fft2(img)
#     f = cv.dft(np.float32(img))
    # 频谱中心化
    fshift = np.fft.fftshift(f)
    res = np.log(np.abs(fshift))
    # 高通滤波
    crow, ccol = int(height/2), int(width/2)
    fshift[crow-filterWidth:crow+filterWidth, ccol-filterWidth:ccol+filterWidth]=0

    #傅里叶逆变
    ishift = np.fft.ifftshift(fshift)
    iimg = np.fft.ifft2(ishift)
#     iimg = cv.idft(ishift)
    iimg = np.abs(iimg)
    
    iimg = np.uint8(iimg)
    out.write(iimg)
    cv.imshow('frame',iimg)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
out.release()
cv.destroyAllWindows()
```
